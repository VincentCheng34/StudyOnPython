```python
# config
proj_path = '/Users/vincent/Desktop/word2Vec'
filePath = '{0}/plaintxt/'.format(proj_path)
fileFullTxt = '{0}/bin/fullTxt_Sentences.txt'.format(proj_path)
# fileFullTxt = '{0}/bin/fullTxt_Paragraph.txt'.format(proj_path)
```

```python
# get segmentated lines from files in folder
from os import walk
import re

def fetchChineseWords(inLines):
    lines = []
    for line in inLines:
        line = re.findall(u'[\u4e00-\u9fff]+', line)
        lines.append(''.join(ch for ch in line))
    return lines

def mergeLines(inLines):
    lines = []
    emptyLine = False
    for line in inLines:
        if line == '':
            emptyLine = True
            continue
        else:
            if emptyLine == True:
                lines[len(lines)-1] = lines[len(lines)-1] + line
            else:
                lines.append(line)
            emptyLine = False
    return lines 
    
def getLinesFromFile_Sentences(filename):
    lines = []
    with open(filename) as fileTrainRaw:
        for line in fileTrainRaw:
            splines = line.split('。');
            for spline in splines:
                if spline != '\n':
                    lines.append(spline)
    return lines

def getLinesFromFile_Paragraph(filename):
    lines = []
    with open(filename) as fileTrainRaw:
        for line in fileTrainRaw:
            lines.append(line)
    return lines

filePureLines = []
def takeKey(elem):
    val = elem.split('.')
    try:
        resInt = int(val[0])
    except ValueError:
        resInt = 0
    return resInt

for (dirpath, dirnames, filenames) in walk(filePath):
    filenames.sort(key=takeKey)
    for filename in filenames:
        if filename == '.DS_Store':
            continue
        fullname = filePath+filename
        print(fullname)
        filePureLines += getLinesFromFile_Sentences(fullname)
#         filePureLines += getLinesFromFile_Paragraph(fullname)
    break

fileTrainSeg = fetchChineseWords(filePureLines)
fileTrainSeg = mergeLines(fileTrainSeg)
# 大雅蒸民天生蒸氏有物有則
fileTrainSeg[620:700]
```

```python
with open(fileFullTxt,'wb') as fW:
    for i in range(len(fileTrainSeg)):
        fW.write(fileTrainSeg[i].encode('utf-8'))
        fW.write('\n'.encode('utf-8'))
```
