## 小於等於兩個字的，視為「分類」

```python
def replaceKey(key):
    return key.replace('　', '').replace(' ', '').replace('\u3000', '').replace('○', '')
```

```python
import re

keywords = []
leftKeywords = []
for idx in range(0, len(allChaptersDf)):
    row = allChaptersDf.loc[idx]
    chapterOri = row[0]
    
    if type(chapterOri) == str:
        chapter = replaceKey(chapterOri)
        if chapter != row[0]:
            print(row[0], chapter)
        if len(chapter) <= 2:
            if chapter not in keywords:
                keywords.append(chapter)
        else:
            if chapter not in leftKeywords:
                leftKeywords.append(chapter)
    else:
        print(chapterOri, type(chapterOri))

print(keywords[0:5])
print(len(keywords))

print(leftKeywords[0:5])
print(len(leftKeywords))
```

## 分割「：」

```python
specialCharList = ['志', '傳']

def replaceSC(key, scList):
    for s in scList:
#         print(len(key));
        if s == key[0] or s == key[-1]:
            return key.replace(s, '')
    return key

def replaceKeyword(key, keyList):
    for k in keyList:
        if k in key:
            return key.replace(k, '')
    return key

def isKeyInKeywords(key, keyList):
    for kl in keyList:
        if kl in key:
#             print(key, kl)
            return True
    return False

def fetchKey(newKey):
    ## 所有冒號：以前的keyword，都已經存在 keyword裡
    if len(newKey) == 2:
        if newKey not in keywords:
            print("new keyword which is from newKey#1 ", newKey)
            keywords.append(newKey)
        return

    #         print(spKeys)
    if newKey not in keywords and len(newKey) > 0:
        key1 = replaceSC(newKey, specialCharList)
        if key1 not in keywords:
            if len(key1) == 2:
                print("new keyword which is from newKey#2 ", key1)
                keywords.append(key1)
            else:
                keyN = replaceKeyword(key1, keywords)
                if len(keyN) == 2:
                    keyM = replaceKeyword(keyN, keywords)
                    if keyN != keyM and len(keyM) == 2:
                        print(key1, keyN, keyM)
    #                 print(firstKey, key1)
    
cnt = 0
for key in leftKeywords:
    if '：' in key:
        cnt += 1
        spKeys = key.split('：')
        firstKey = replaceKey(spKeys[0])
        secondKey = replaceKey(spKeys[1])

        fetchKey(firstKey)
        fetchKey(secondKey)

print(len(keywords))
print(cnt)
```

```python
leftKeywords2 = []
for lkey in leftKeywords:
    nlkey = replaceKey(lkey)
    for key in keywords:
        if key in nlkey:
            break
#     print(key, lkey)
    if key not in nlkey:
        leftKeywords2.append(nlkey)
        
leftKeywords = leftKeywords2
print(len(leftKeywords))
```
