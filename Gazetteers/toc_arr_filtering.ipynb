## filtering special characters

```python
import re
# regex = re.compile(r'([卷一首二三四五六七八九十之苐第]+)')

def hasSpecialWord(INword, specialWords):
    for w in specialWords:
        if w in INword:
            return True
    return False

def removeRegex(INdict, INmatchLen):
    # specialWords = ['卷', '一', '首', '二', '三', '四', '五', '六', '七', '八', '九', '十', '之', '苐', '第']
    regex = re.compile(r'([帙至卷一首二三四五六七八九十廿卄之苐第上中下]+)')
    passList = '科'
    
    removeList = {}
    for c in INdict:
        subArr = INdict[c]
        for sub in subArr:
            if passList in sub:
                continue

            match = regex.search(sub)

            if match:
                start, end = match.span()
                if end - start >= INmatchLen:
                    if start != 0 and end != len(sub):
                        print(sub, match)
#                         removeList[sub] = ''
                        continue
#                     print(sub, match)
                    
                    matchValue = sub[start:end]
                    newValue = sub.replace(matchValue, '')
                    removeList[sub] = newValue
                    
    return removeList

removeObjs = removeRegex(OriDict, 1)
removeObjs
```

```python
def removeValues(INobjs, INremoveObjs):
    newCategory = {}
    for key in INobjs.keys():
        subArr = INobjs[key]
        
        newCategory[key] = []
        for sub in subArr:
            if sub in INremoveObjs:
#                 print(sub)
                newValue = INremoveObjs[sub]
                if len(newValue) > 0 and newValue not in newCategory:
#                     print(sub, newValue)
                    newCategory[key].append(newValue)

            else:
                newCategory[key].append(sub)
            
    return newCategory

OriDict = removeValues(OriDict, removeObjs)
# OriDict
```
